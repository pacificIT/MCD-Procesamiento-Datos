
Se realizo la eliminacion del indice y demas texto que siempre es igual para poder acceder al texto un poco mas relevane, en este caso aparir del prefacio. 


```python
# se optiene el texto del archivo 

soup = BeautifulSoup(austen)

text = soup.get_text()
```

```python
# Se visualiza una muestra de texto a procesar

print(text[0:3000])
```

```python
# se calcula el tamaño para tener una referencia del mismo 

len(text)
```

```python
# se realiza a busqueda de tokens de la cadena utilizando una expresión regular con el RegexpTokenizer() 

tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(text)
```

```python
# se obtienen N tokens , es decir palabras 

len(tokens)
```

```python
# tokens iniciales 
print(tokens[0:200])
```

```python
# todo se convierte a minuscula los tokens (palabras) para un procesamiento  uniforme 

words = [word.lower() for word in tokens]

print(words[:200])
```

```python
# se obtiene la lista de stop words para ayudar en procesamiento
sw = nltk.corpus.stopwords.words('english')
print(sw[:50])
```

```python
# se obtienen las palabras en el texto que no son stop words
words_ns = [word for word in words if word not in sw]
print(words_ns[:250])
```

# Lematización


La lematización es el proceso de agrupar las diferentes formas flexionadas de una palabra para que puedan analizarse como un solo elemento. La lematización es similar a la derivación, pero aporta contexto a las palabras. Por lo tanto, vincula palabras con un significado similar a una palabra. la lematización realiza un análisis morfológico de las palabras.

Se utiliza con frecuencia en :

    Sistemas de recuperación integrales como motores de búsqueda.
    
    Indexación compacta


```python
# cargar lematizado 
lemmatizer = WordNetLemmatizer() 
print("rocks :", lemmatizer.lemmatize("rocks")) 
print("corpora :", lemmatizer.lemmatize("corpora")) 
# a denotes adjective in "pos" 
print("miss :", lemmatizer.lemmatize("miss", pos ="a"))
```

```python
# lematizado sin stop words 
words_lem = [lemmatizer.lemmatize(word) for word in words_ns]
print(words_lem[:250])
```

```python
# Graficando la frecuencia de las palabras lamatizadas 
from matplotlib.pyplot import figure
%matplotlib inline

# se calcula la frecuencia de las palabras
freqdist = nltk.FreqDist(words_ns)
```

```python
freqdist
```

```python
# Plotting the word frequency distribution
figure(figsize=(15,5))
freqdist.plot(50)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(words_ns)
```
